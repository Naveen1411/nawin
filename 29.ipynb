```python
#Calculate the average rating of Gaming and Non-Gaming apps
sum=0
for i in f.user_rating_ver:
    sum+=i
print("average_rating for paid=",sum/len(f.user_rating_ver))
```

```python
#Calculate the average rating of free Gaming apps
a=x[x.prime_genre=='Games']
add=0
for i in a.user_rating_ver:
    add+=i
print("average rating for free gaming=",add/len(a.user_rating_ver))
```

```python
#Compute the average rating of the apps whose genre is either "Social Networking" or "Games."
b=f[(f.prime_genre=='Social Networking') | (f.prime_genre=="Games")]
sum=0
for  i in b.user_rating_ver:
    sum+=i
print("average rating of the apps whose genre is eitherSocial Networking or Games=",sum/len(b.price))
```

```python
#Compute the average rating of the non-free apps whose genre is either "Social Networking" or "Games."
d=y[(y.prime_genre=='Social Networking') | (y.prime_genre=="Games")]
sum=0
for  i in d.user_rating_ver:
    sum+=i
print("average rating of the non free apps whose genre is eitherSocial Networking or Games=",sum/len(d.price))
```

```python
#Compute the average rating of the apps that have a price greater than $9.
e=f[f.price>=9]
sum=0
for i in e.user_rating_ver:
    sum+=i
print("Compute the average rating of the apps that have a price greater than $9=",sum/len(e.price))
```

```python
#Categorise all apps by labelling each app as "free"(=0),  "affordable" (<20), "expensive" (<50) or "very expensive" (>50). Add a label column to the data.
import pandas as pd
l=[]
for i in f.price:
    if i==0:
        l.append('free')
    elif i>0 and i<20:
        l.append('affordable')
    elif i>=20 and 1<50:
        l.append("expensive")
    else:
        l.append("veryexpensive")
print(len(l))
s=pd.DataFrame(l)
s.rename(columns={0:"label"})
w=pd.concat([s,f])
w




```
