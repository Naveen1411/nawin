```
#Write a python program to find the longest words
f=open("file1.txt","w")
f.write("this is naveen kumar reddy")
f=open("file1.txt","r")
l=f.readlines()
for i in l:
  list=i.split(" ")
print(list)
x=max(len(x) for x in list)
print(x)
for a in list:
  n=a
  if len(a)==x:
    print(n)
```

```
#Write a python program to count the character from given file.
f=open("file1.txt","w")
f.write("this is naveen kumar reddy")
f=open("file1.txt","r")
x=f.readline()
c=0
for ch in x:
  if ch.isalpha():
    c=c+1
print(c)
```

```
#Write a Python program to count the frequency of words in a file
f=open("file.txt","w")
f.write("hello world")
f=open("file.txt","r")
s=f.read()
d={}
for i in s:
  if i in d:
    d[i]+=1
  else:
    d[i]=1
print(d)
    
```

```
#Write a Python program to copy the contents of a file to another file
f1=open("file1.txt","w")
f1.write("dhdfdgj fnkslj\nbdjbhodbfhboj\njfghfdso")
f1=open("file1.txt","r")
x=f1.read()
print(x)
f2=open("file2.txt","w")
for line in x:
  
  f2.write(line)
f2=open("file2.txt","r")
print(f2.read())
f1.close()
```
