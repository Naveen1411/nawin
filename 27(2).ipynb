```python
import matplotlib.pyplot as pt
a=[1,2,3,4]
b=[1,2,3,4]
pt.figure(figsize=(4,5))
pt.plot(a,b,'g-.>')
pt.title("bar graph")
pt.xlabel('time')
pt.ylabel('x(t)')
pt.show()
```

```python
import numpy as np
x=np.arange(1,6)
y=x+1
pt.plot([6,5,4,3,2,1],[1,2,3,4,5,6],'b.-',x,y,'g-.^')
pt.title("dual graph")
pt.xlabel("x")
pt.ylabel("y")
pt.figure(figsize=(6,6))
pt.show()
```

```python
a=['a','b','c','d']
b=[25,22,23,15]


index=np.arange(len(a))
pt.barh(a,b,color="g")
```

```python
a=['a','b','c','d']
b=[25,22,23,15]
c=[1,5,3,1]
width=0.5

index=np.arange(len(a))
pt.barh(a,b,color="g",label='l1')
pt.barh(index+width,c,color="r",label='l2')
pt.legend()
```

```python
index = np.arange(len(a))
width = 0.3
pt.bar(a,b, color = "#1412cd", width = 0.3,label='bar2')
pt.bar(index+width, b, color = ['red', 'g'], width = 0.3, label = "bar1")
pt.legend()
```

```python
l=['india','england','australia','newzland']
w=[60,20,15,5]
pt.pie(w,labels=l,explode=[0,0.2,0,0],shadow=True,autopct='%1.1f%%')
pt.legend(loc="best")
```
