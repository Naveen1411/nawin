```python

import pandas as pd
f=pd.read_csv(r'D:\AppleStore.csv')
f
```

```python
#Compute the total number of unique apps from the datase
l=[]
for x in f.track_name:
    if x not in l:
        l.append(x)
print(l)
print("-"*10)
print(len(l))
        

    
```

```python
#Print the Top 10 apps along with their rating based on the number of downloads(rating_count_tot)
l=[]
for i in f.rating_count_tot:
    l.append(i)

#(len(l))
l.sort(reverse=True)
#print(l)
for i in range(10):
    print(l[i]) 
```

```python
#Categorise the dataset based on content rating into the following
a=0
b=0
c=0
d=0
w=[]
for i in f.cont_rating:
    if i=="4+":
        a+=1
    elif i=="9+":
        b+=1
    elif i=="12+":
        c+=1
    elif i=="17+":
        d+=1
    else:
        pass
w.append(a)
w.append(b)
w.append(c)
w.append(d)
print("Number of apps with content rating 4+:",a)
print("Number of apps with content rating 9+:",b)
print("Number of apps with content rating 12+:",c)
print("Number of apps with content rating 17+:",d)
print(w)
```

```python
#Accessing the App data from App Store, generate the following Visualizations
#Pie Chart to categorise all app categories based on count
import matplotlib.pyplot as pt
l=[]
for x in f.cont_rating:
    if x not in l:
        l.append(x)
print(l,w)
pt.pie(w,labels=l,autopct='%1.1f%%')
pt.legend(loc="best")
        
```
